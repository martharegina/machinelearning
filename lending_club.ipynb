```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
data = pd.read_csv(
    '/content/accepted_2007_to_2018Q4.csv.gz',
    compression='gzip',
    low_memory=False)
data.head()
```

```python
object_cols = [
    'annual_inc',
    'dti',
    'fico_range_low',
    'fico_range_high',
    'emp_length',
    'revol_util',
    'total_acc',
    'open_acc',
    'delinq_2yrs',
    'loan_amnt',
    'installment',
    'term',
    'int_rate',
    'purpose',
    'application_type',
    'home_ownership',
    'issue_d'
]

data = data[object_cols]
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data = data.dropna(axis=0)
data.shape
```

```python
data.info()
```

```python
# Pilih data di tahun 2018 saja

data['issue_d'] = pd.to_datetime(data['issue_d'])
data = data[data['issue_d'].dt.year == 2018]
data.info()
```

Penjelasan setiap kolom:
* annual_inc: pendapatan tahunan peminjam, ukuran utama kemampuan bayar
* dti (debt-to-income ratio): persentase total cicilan terhadap pendapatan
* fico_range_low dan fico_range_high: skor kredit (batas bawah dan atas)
* emp_length: lama bekerja (stabilitas pendapatan)
* revol_util: persentase pengunaan kredit bergulir (kartu kredit)
* total_acc: total jumlah akun kredit sepanjang hidup
* open_acc: jumlah akun kredit aktif
* delinq_2yrs: jumlah keterlambatan >30 hari dalam 2 tahun terakhir
* loan_amt: jumlah pinjaman aktual
* installment: cicilan bulanan
* term: tenor (36 atau 60 bulan)
* int_rate: suku bunga
* purpose: tujuan pinjaman
* application_type: individual atau joint
* home_ownership: status kepemilikan rumah (rent, own, mortgage)
* issue_d: tanggal pinjaman diterbitkan

```python
# Ordinal encoding emp_length
emp_map = {
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10
}
data['emp_length'] = data['emp_length'].map(emp_map)
data.head()
```

```python
# Binary encoding term
data['term'] = data['term'].map({' 36 months': 0, ' 60 months': 1})
data.head()
```
